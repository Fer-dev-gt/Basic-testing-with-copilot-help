```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://techcrunch.com/'

response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

posts = soup.find_all('div', class_='post-block post-block--image post-block--unread')

data = []

for post in posts:
    title = post.find('h2', class_='post-block__title').text.strip()
    link = post.find('a', class_='post-block__title__link')['href']
    date = post.find('time', class_='river-byline__time')['datetime']
    author = post.find('span', class_='river-byline__authors').text.strip()
    content = post.find('div', class_='post-block__content').text.strip()

    data.append({
        'title': title,
        'link': link,
        'date': date,
        'author': author,
        'content': content,
    })

df = pd.DataFrame(data)
```

```python
# escribe una funcion para separar la fecha de la hora en dos columnas diferentes, de la columna "date"

def split_date(date):
    date = date.split('T')
    return date[0], date[1]

df['date'], df['time'] = zip(*df['date'].map(split_date))
```

```python
## Crea una funcion para eliminar duplicados y reemplazar los valores nulos por 'N/A' en el dataframe df 

def clean_df(df):
    df = df.drop_duplicates()
    df = df.fillna('N/A')
    return df

df = clean_df(df)
```
